# Matplotlib Tutorail

## Load Necessary Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Basic Graph

```python
x = [0, 1, 2, 3, 4]
y = [0, 2, 4, 6, 8]

plt.plot(x, y, label='2x', color='red', linewidth=3, linestyle='--', marker='.', markersize=15,markeredgecolor='blue')

plt.title("Our first Graph!", fontdict={'fontname': 'Comic Sans Ms', 'fontsize': 20})
plt.xlabel("X Axis", fontdict={"fontname": 'Comic Sans MS'})
plt.ylabel("Y Axis")
plt.xticks([0,1,2,3,3.5,4])
plt.yticks([0,2,4,6,8,10])
plt.legend()

plt.show()
```

# Format Strings

A format string consists of a part for color, marker and line:
    
    fmt = '[marker][line][color]'


Each of them is optional. If not provided, the value from the style
cycle is used. Exception: If `line` is given, but no `marker`
the data will be a line without markers.

Other combinations such as `[color][marker][line]` are also
supported, but note that their parsing may be ambiguous.

```python

x = [0, 1, 2, 3, 4]
y = [0, 2, 4, 6, 8]

# Resize your graph
plt.figure(figsize=(5, 3), dpi=100)

# Plot x and y
plt.plot(x, y, 'b^--', label='2x')

#Line number 2
x2 = np.arange(0,4.5,0.5)

# plot part of the graph
plt.plot(x2[:6], x2[:6]**2, 'r', label='x^2')

# plot remaining part of graph as a dot
plt.plot(x2[5:], x2[5:]**2, 'r--')

# plot title (specify font parameter with fontdict)
plt.title("Our first Graph!", fontdict={'fontname': 'Comic Sans Ms', 'fontsize': 20})

# X label
plt.xlabel("X Axis", fontdict={"fontname": 'Comic Sans MS'})

# y label
plt.ylabel("Y Axis")

# x, y tickmarks (Scale your graph)
plt.xticks([0,1,2,3,3.5,4])

# graph legend
plt.legend()

# save your figure (dpi 300 is good when saving so graph has high resolution)
plt.savefig('my_graph.png', dpi=300)

# show plot
plt.show()
```

## Bar Chart

```python
labels = ['A', 'B', 'C']
values = [1,4,2]

plt.figure(figsize=(6,4))
bars = plt.bar(labels, values)

bars[0].set_hatch('/')
bars[1].set_hatch('o')
bars[2].set_hatch('*')



plt.show()
```

```python
labels = ['A', 'B', 'C']
values = [1,4,2]

plt.figure(figsize=(6,4))
bars = plt.bar(labels, values)
patterns = ['/', '*', 'o']
for bar in bars:
    bar.set_hatch(patterns.pop())


plt.show()
```

## Real World Examples

### Line Graph

```python
gas = pd.read_csv('gas_prices.csv')
gas
```

```python
# Line plot

# figure size
plt.figure(figsize=(8,5))

# plot
plt.plot(gas.Year, gas.USA, 'b.-', label='USA')
plt.plot(gas['Year'], gas.Canada, 'r.-', label='Canada')
plt.plot(gas['Year'], gas['South Korea'], 'g.-', label='South Korea')
plt.plot(gas.Year, gas['Australia'], 'y.-', label='Australia')

# Add title
plt.title("Gas Prices over time (in USD)", fontdict={"fontweight": 'bold'})

# Axis labels
plt.xlabel("Year")
plt.ylabel('Us Dollars') 

# plot tickmarks
plt.xticks(gas.Year[::2])

# Add legend
plt.legend()

# show plot
plt.show()
```

```python
# Line plot

# figure size
plt.figure(figsize=(8,5))

# plot

for country in gas.iloc[:, 1:]:
    plt.plot(gas.Year, gas[country], marker='.', label=country)

# Add title
plt.title("Gas Prices over time (in USD)")

# plot tickmarks
plt.xticks(gas.Year[::2])

# Axis labels
plt.xlabel("Year")
plt.ylabel('Us Dollars')

# Add legend
plt.legend()

# show plot
plt.show()
```

```python
fifa = pd.read_csv('fifa.csv')
fifa
```

## Histograms

```python
bins = [30, 40 , 50, 60, 70, 80, 90, 100]
plt.hist(fifa.Overall, bins=bins, color='lime')

plt.xticks(bins)

plt.xlabel("Skill Level")
plt.ylabel("Number of players")
plt.title("Distribution of player skill in FIFA 2018")

# Show plot
plt.show()
```

## Pie Chart

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

plt.pie([left, right], labels=['Left', 'Right'], colors=['red', 'blue'], autopct="%.2f %%")

plt.title("Foot Preference of players")

plt.show()
```

```python
# Pie chart #2
fifa.Weight = [int(x.strip('lbs')) if type(x) == str else x for x in fifa.Weight]

light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa.loc[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa.loc[fifa.Weight >= 200].count()[0]

weights = [light, light_medium, medium, medium_heavy, heavy]
labels = ["Under 125", "125-150", "150-175", "175-200", "Over 200"]
explode = (.4,.2, 0, 0, .4)

#plt.style.use('ggplot')
plt.figure(dpi=200)
plt.pie(weights, labels=labels, autopct="%.2f %%", pctdistance=0.7, explode=explode)

plt.title("Weight Distribution of FIFA Players (in lbs)")

plt.show()
```

## Box Plots

```python
plt.style.use('default')
plt.figure(figsize=(5,8))

barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
revs = fifa.loc[fifa.Club == 'New England Revolution']['Overall']

boxes = plt.boxplot([barcelona, madrid, revs],
    labels=['FC Barcelona', 'Real Madrid', 'NE Revolution'],
    patch_artist=True,
    medianprops={'linewidth':2})

for box in boxes['boxes']:
    # set edge color
    box.set(color='#4286f4', linewidth=2) 

    # change fill color
    box.set(facecolor='#e0e0e0')

plt.title("Professional Soccer Team Comparison")
plt.ylabel("FIFA Overall Rating")

plt.show()
```
